    sim.d[t,:] = np.clip(tools.interp_2d_vec(sol.grid_d_old[t,:], sol.grid_w_old[t,:], sol.d[t,:,:, 0], sim.d_old[t,:], sim.w_old[t,:]), a_min = 0, a_max=0.74)

        if credit_con <= 0.05:
            sim.d_old[t+1,:] = np.minimum((1-par.lambdaa)*sim.d_old[t,:], sim.d[t,:])
