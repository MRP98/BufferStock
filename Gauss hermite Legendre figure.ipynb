```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = 'garamond'

n = 25  # number of nodes
quad_methods = [np.polynomial.legendre.leggauss, np.polynomial.hermite.hermgauss]
colors = ['red', 'blue']
label = ["Gauss-Legendre $[-1,1]$", "Gauss-Hermite  $[-1,1]$"]
fig, ax = plt.subplots(figsize=(8, 5))

for i, method in enumerate(quad_methods):
    x, w = quad_methods[i](n)
    ax.scatter(x, w, label=label[i], color=colors[i])
    
# Plot the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation
x_values = np.linspace(-3.5, 3.5, 100)
y_values = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x_values - mu) / sigma) ** 2)
ax.plot(x_values, y_values, color='gray', linestyle='--', label='Normal Distribution')

ax.set_title('Quadrature Nodes and Normal Distribution', fontsize=18)
ax.legend(fontsize=12)

plt.show()
```
